```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("doctor_appointment.db")
cursor = conn.cursor()

# Create Patients table
cursor.execute('''CREATE TABLE IF NOT EXISTS Patients (
                    patient_id INTEGER PRIMARY KEY,
                    name TEXT,
                    age INTEGER,
                    gender TEXT,
                    email TEXT UNIQUE
                 )''')

# Create Doctors table
cursor.execute('''CREATE TABLE IF NOT EXISTS Doctors (
                    doctor_id INTEGER PRIMARY KEY,
                    name TEXT,
                    specialty TEXT,
                    available_days TEXT
                 )''')

# Create Appointments table
cursor.execute('''CREATE TABLE IF NOT EXISTS Appointments (
                    appointment_id INTEGER PRIMARY KEY,
                    patient_id INTEGER,
                    doctor_id INTEGER,
                    appointment_date TEXT,
                    status TEXT CHECK(status IN ('Booked', 'Rescheduled', 'Cancelled')),
                    FOREIGN KEY (patient_id) REFERENCES Patients(patient_id),
                    FOREIGN KEY (doctor_id) REFERENCES Doctors(doctor_id)
                 )''')

conn.commit()
```

```
# Insert sample patients
cursor.execute("INSERT INTO Patients (name, age, gender, email) VALUES ('Alice Johnson', 28, 'Female', 'alice@email.com')")
cursor.execute("INSERT INTO Patients (name, age, gender, email) VALUES ('Bob Smith', 35, 'Male', 'bob@email.com')")

# Insert sample doctors
cursor.execute("INSERT INTO Doctors (name, specialty, available_days) VALUES ('Dr. John Doe', 'Cardiologist', 'Monday, Wednesday, Friday')")
cursor.execute("INSERT INTO Doctors (name, specialty, available_days) VALUES ('Dr. Sarah Lee', 'Dermatologist', 'Tuesday, Thursday')")

conn.commit()
```
